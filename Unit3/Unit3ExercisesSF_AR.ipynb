## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english.

Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.


*   **Bernoulli:** Supports x values of 0 and 1. This distribution models the probability of a single trial with two possible outcomes: success (represented by 1) with probability *p*, and failure (represented by 0) with probability 1-*p*.

*   **Binomial:** Supports integer x values from 0 to *n*. This distribution models the number of successes in a fixed sequence of *n* independent Bernoulli trials, where each trial has a probability of success *p*.

*   **Poisson:** Supports non-negative integer x values (0, 1, 2, ...). This distribution models the number of independent events occurring in a fixed interval of time or space, given a known average rate of occurrence (μ).

*   **Geometric:** Supports positive integer x values (1, 2, 3, ...). This distribution models the number of Bernoulli trials needed to get the first success, where each trial has a probability of success *p*.

*   **HyperGeometric:** Supports integer x values from max(0, n - (N - k)) to min(n, k). This distribution models the number of successes in a fixed number of *n* draws from a finite population of size *N* without replacement, where *k* items in the population are successes.

*   **Categorical:** Supports non-negative integer x values from 0 up to length(p) - 1. This distribution models the probability of an outcome from a set of discrete outcomes, where each outcome has a specified probability given by the vector p.
